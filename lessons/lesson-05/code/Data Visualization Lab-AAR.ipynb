```python
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns

# Make plots larger
plt.rcParams['figure.figsize'] = (10, 6)
```

First let's try a some function plotting.

```python
# Plot two normal distributions

domain = np.arange(-20, 20, 0.1)
# np.arange returns an ndarray (not a list) of evenly spaced values within a given interval and calls it 'domain'. 

values = stats.norm(3, 5).pdf(domain)
# creates a normal continuous random variable, with mean=3 and std=5, pdf is probability density function 

plt.plot(domain, values, color='r', linewidth=2)
# creates plot with "domain" as x, "values" as y, line color is red, and linewidth is 2

plt.fill_between(domain, 0, values, color='r', alpha=0.3)
# fills in the color red for the bottom graph

values = stats.norm(4, 2).pdf(domain)
# creates a normal continuous random variable, with mean=4 and std=2, pdf is probability density function

plt.plot(domain, values, color='b', linewidth=2)
# creates plot with "domain" as x, "values" (this is the second  'values' variable defined) as y, 
#line color is blue, and linewidth is 2

plt.ylabel("Probability")
# adds label to y axis

plt.title("Two Normal Distributions")
#adds title

plt.show()
# displays the figure as defined above
```

# Time Series plots

Matplotlib and Seaborn can make some nice plots associated to time series data. For example, we can make plots of running. The following data contains the monthly price of the ETF VTI (a stock market index fund) over time

```python
df = pd.read_csv(os.path.join("datasets", "vti.csv"))
# reads in dataset from pathname joined by os

df.dtypes
```

```python
df = pd.read_csv(os.path.join("datasets", "vti.csv"))
# reads in dataset from pathname joined by os

df.sort_values(by="Date", inplace=True)
# sorts dataset by date column, 
# inplace=True means the current dataset is modified rather than making a new dataset with the column sorted 

df["Date"] = pd.to_datetime(df["Date"], format='%Y-%m-%d')
#converts an 'object' datatype to a 'datetime' datatype

df.head()
```

```python
#plt.plot seems to plot a line chart by default

plt.plot(df["Date"], df["Open"], label="Open")
# first chart series, date is on x-axis and open is on y-axis

plt.plot(df["Date"], df["Close"], label="Close")
# first chart series, date is on x-axis and close is on y-axis

plt.xlabel("Date")
#sets label for x-axis

plt.ylabel("Price")
#sets label for y-axis

plt.title("VTI Monthly Prices")
#sets title for plot

plt.legend()
#adds legend to plot

plt.show()
#actually shows the plot
```

```python
#Plot to show difference between close and open over time vs. the difference between high and low over time.

plt.plot(df["Date"], df["Open"] - df["Close"], label="Close-Open")
plt.plot(df["Date"], df["High"] - df["Low"], label="High-Low")
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("VTI Monthly Prices")
plt.legend()
plt.show()
```

## Error bars and filled plots

Often we want to indicate that our data is noisy or contains measurement error. Let's construct a dataset.

```python
# Check: do you understand this code?
import numpy as np
from scipy import stats
import random
data = []
for i in range(50):
    m = random.randint(5 + i, 15 + i)
    s = random.randint(4, 8)
    dist = stats.norm(m, s)
    draws = dist.rvs(30)
    data.append([np.mean(draws), np.std(draws)])
df = pd.DataFrame(data, columns=["Mean", "Std"])
df.head()
```

```python
# Now we can plot with error bars
plt.errorbar(range(len(df)), df["Mean"], yerr=df["Std"])
plt.title("Error Bar Example")
```

```python
# Confidence interval: 68%
plt.errorbar(range(len(df)), df["Mean"])
lower = df["Mean"] - df["Std"]
upper = df["Mean"] + df["Std"]
plt.fill_between(range(len(df)), lower, upper, alpha=0.5)
plt.title("CI Example")
```
