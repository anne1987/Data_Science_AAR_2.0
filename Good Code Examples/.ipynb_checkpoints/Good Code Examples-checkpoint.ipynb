```python
#C:\Users\anne1\Documents\DS_SM_15\Good Code Examples
#C:\Users\anne1\Documents\DS_SM_15\lessons\lesson-06\code\dataset\msleep

#Load mammals dataset

#creates pathname object for where the file is saved
wd = '../lessons/lesson-06/code/dataset/msleep/'

#creates a dataframe object called mammals that is created from reading in a csv file, whose full name is 
#is created by concatenating the pathname object 'wd' and the file name 
mammals = pd.read_csv(wd+'msleep.csv')
```

```python
#shows first row of dataset
mammals.head(1)
```

```python
# get the mean of the sleep_total columns for each order
mammals.groupby('order')['sleep_total'].mean()
```

```python
# rows are grouped by order and then genus, then the mean of their 'sleep_total' is returned
# order matters, be sure that the larger category is listed before the sub-category
mammals.groupby(['order', 'genus'])['sleep_total'].mean()
```

```python
# Groups by Order, then genus.  Then calculates the mean for the variables 'sleep_total' and 'awake'.
mammals.groupby(['order', 'genus'])['sleep_total','awake'].mean()
```

```python
mammals.head()
```

```python
# rows grouped by order and genus. Columns for sleep_total and awake, 
#Agg function to apply distinct list of computations to each column
# Agg function takes a dict where the column name is the key and the computations are the list
mammals.groupby(['order', 'genus'])['sleep_total','awake'].agg({
    'sleep_total':['sum','mean','max'],
    'awake': ['max','min','sum']})
    
```
